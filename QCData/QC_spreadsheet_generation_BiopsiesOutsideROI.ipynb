```r
## using fourth parse b/c that's w/ no screenshots: 
recgli = read.csv("../ParseData/4thParse_researchPath_withInVivo.csv")
```

```r
colnames(recgli)
```

```r
recgli.desiredhist = recgli[recgli$desired_hist==1,]
```

```r
dim(recgli.desiredhist)
```

```r
recgli.nonecrosis = recgli.desiredhist[-which(recgli.desiredhist$necrosis==2),]
recgli.nonecrosis = recgli.nonecrosis[-which(recgli.nonecrosis$in_NEC == 1),]
```

```r
dim(recgli.nonecrosis)
```

```r
recgli = recgli.nonecrosis
```

```r
table(recgli$in_ROI)
```

```r
recgli_outsideROI = recgli[which(recgli$in_ROI ==0),]
```

```r
dim(recgli_outsideROI)
```

```r
recgli = recgli_outsideROI
recgli = recgli[order(recgli$t_number),]
recgli
```

```r
## max number of biopsies in the same t_number seems to be four 
#vialid_df = data.frame(t_number = recgli$t_number[!duplicated(recgli$t_number)], 
#                      vialid1=NA, vialid2=NA, vialid3=NA, vialid4=NA)
```

```r
qc = data.frame(b_number  = recgli$b_number, t_number = recgli$t_number, roi.label = recgli$roi.label, olddata = recgli$olddata, 
               tumor_cell_evaluation = recgli$tumor_cell_evaluation, necrosis = recgli$necrosis, nfse = recgli$nfse, 
                nfl = recgli$nfl, nt1c = recgli$nt1c, nt1v=recgli$nt1v, X.CEL = recgli$X.CEL, X.NEL = recgli$X.NEL, X.NEC = recgli$X.NEC ) 

qc = data.frame(qc, adc_align_check = NA, firsta_align_check = NA, first_check = NA, cbv_map_check = NA, 
               cel_qc = NA, t2all_qc = NA, nel_qc = NA, necrosis_presence_check = NA, nec_qc = NA, 
               biopsies_outside_rois_check = NA, notes = NA)
```

```r
head(qc)
```

```r
qc = qc[order(qc$t_number),]
```

```r
write.csv(qc, file = "qc_biopsies_outside_roi_template.csv", row.names =F)
```
