## 1. Anatomical vs. rHGG/TxE

### a. No separation into ROIs

```r
par(mfrow = c(2, 2))
a = boxplot(nfse ~ multnom_out , data = recgli, ylab = "normalized fse", main = "median fse")
text( c(1:nlevels(factor(recgli$multnom_out))) , a$stats[nrow(a$stats) , ]+0.5 , paste("n = ",table(recgli$multnom_out),sep="")  )

b = boxplot(nfl ~ multnom_out , data = recgli, ylab = "normalized fl", main = "median fl")
text( c(1:nlevels(factor(recgli$tumor_cell_evaluation))) , b$stats[nrow(b$stats) , ]+0.5 , paste("n = ",table(recgli$multnom_out),sep="")  )

c = boxplot(nt1c ~ multnom_out , data = recgli, ylab = "normalized t1c", main = "median t1c")
text( c(1:nlevels(factor(recgli$multnom_out))) , c$stats[nrow(c$stats) , ]+0.5 , paste("n = ",table(recgli$multnom_out),sep="")  )

d = boxplot(nt1v ~ multnom_out , data = recgli, ylab = "normalized t1v", main = "median t1v")
text( c(1:nlevels(factor(recgli$multnom_out))) , d$stats[nrow(d$stats) , ]+0.5 , paste("n = ",table(recgli$multnom_out),sep="")  )
```

### b. CEL next to NEL scores  

```r
par(mfrow = c(2, 2))
a = boxplot(nfse ~ multnom_out , data = recgli[recgli$in_CEL ==1,], ylab = "normalized fse", main = "CEL median FSE", ylim = c(0,4))
text( c(1:nlevels(factor(recgli$multnom_out))) , a$stats[nrow(a$stats) , ]+.3 , paste("n = ",table(recgli[which(recgli$in_CEL ==1),]$multnom_out),sep=""), cex = 0.8)

b = boxplot(nfse ~ multnom_out , data = recgli[recgli$in_NEL ==1,], ylab = "normalized fse", main = "NEL median FSE", ylim = c(0,4))
text( c(1:nlevels(factor(recgli$multnom_out))) , b$stats[nrow(b$stats) , ]+.3 , paste("n = ",table(recgli[which(recgli$in_NEL ==1),]$multnom_out),sep=""), cex = 0.8)

c = boxplot(nfl ~ multnom_out , data = recgli[recgli$in_CEL ==1,], ylab = "normalized fl", main = "CEL median NFL", ylim = c(0,3.5))
text( c(1:nlevels(factor(recgli$multnom_out))) , c$stats[nrow(c$stats) , ]+.3 , paste("n = ",table(recgli[which(recgli$in_CEL ==1),]$multnom_out),sep=""), cex = 0.8)
d = boxplot(nfl ~ multnom_out , data = recgli[recgli$in_NEL ==1,], ylab = "normalized fl", main = "NEL median NFL", ylim = c(0,3.5))
text( c(1:nlevels(factor(recgli$multnom_out))) , d$stats[nrow(d$stats) , ]+.3 , paste("n = ",table(recgli[which(recgli$in_NEL ==1),]$multnom_out),sep=""), cex = 0.8)

e = boxplot(nt1c ~ multnom_out , data = recgli[recgli$in_CEL ==1,], ylab = "normalized t1c", main = "CEL median T1c", ylim = c(0,4))
text( c(1:nlevels(factor(recgli$multnom_out))) , e$stats[nrow(e$stats) , ]+.3 , paste("n = ",table(recgli[recgli$in_CEL ==1,]$multnom_out),sep=""), cex = 0.8)
f = boxplot(nt1c ~ multnom_out , data = recgli[recgli$in_NEL ==1,], ylab = "normalized t1c", main = "NEL median T1c", ylim = c(0,4))
text( c(1:nlevels(factor(recgli$multnom_out))) , f$stats[nrow(f$stats) , ]+.3 , paste("n = ",table(recgli[recgli$in_NEL ==1,]$multnom_out),sep=""), cex = 0.8)

g = boxplot(nt1v ~ multnom_out , data = recgli[recgli$in_CEL ==1,], ylab = "normalized t1v", main = "CEL median T1v", ylim = c(0,4))
text( c(1:nlevels(factor(recgli$multnom_out))) , g$stats[nrow(g$stats) , ]+.3 , paste("n = ",table(recgli[recgli$in_CEL ==1,]$multnom_out),sep=""), cex = 0.8)
h = boxplot(nt1v ~ multnom_out , data = recgli[recgli$in_NEL ==1,], ylab = "normalized t1c", main = "NEL median T1v", ylim = c(0,4))
text( c(1:nlevels(factor(recgli$multnom_out))) , h$stats[nrow(h$stats) , ]+.3 , paste("n = ",table(recgli[recgli$in_NEL ==1,]$multnom_out),sep=""), cex = 0.8)
```

```r
nt1v.gee.fit = geeglm(binom_out ~ nt1v, data = recgli[!is.na(recgli$nt1v),], id = factor(t_number), family = binomial, corstr = "exchangeable")
round(summary(nt1v.gee.fit)$coefficients, 4)
```

### e. Use GLMM to associate anatomical parameters with TxE/rHGG, separating into CEL & NEL 

```r
# NFSE 
nfse.cel.glmm.fit = glmer(multnom_out ~ nfse + (1|t_number), data = recgli[which(recgli$in_CEL ==1),], family="binomial")
nfse.nel.glmm.fit = glmer(multnom_out ~ nfse + (1|t_number), data = recgli[which(recgli$in_NEL ==1),], family="binomial")
print("In CEL:")
round(summary(nfse.cel.glmm.fit)$coefficients, 4)
print("In NEL:")
round(summary(nfse.nel.glmm.fit)$coefficients, 4)
```

```r
# NFL 
nfl.cel.glmm.fit = glmer(multnom_out ~ nfl + (1|t_number), data = recgli[which(recgli$in_CEL ==1),], family="binomial")
nfl.nel.glmm.fit = glmer(multnom_out ~ nfl + (1|t_number), data = recgli[which(recgli$in_NEL ==1),], family="binomial")
print("In CEL:")
round(summary(nfl.cel.glmm.fit)$coefficients, 4)
print("In NEL:")
round(summary(nfl.nel.glmm.fit)$coefficients, 4)
```

```r
# NT1c
nt1c.cel.glmm.fit = glmer(multnom_out ~ nt1c + (1|t_number), data = recgli[which(recgli$in_CEL ==1),], family="binomial")
nt1c.nel.glmm.fit = glmer(multnom_out ~ nt1c + (1|t_number), data = recgli[which(recgli$in_NEL ==1),], family="binomial")
print("In CEL:")
round(summary(nt1c.cel.glmm.fit)$coefficients, 4)
print("In NEL:")
round(summary(nt1c.nel.glmm.fit)$coefficients, 4)
```

```r
# NT1v
nt1v.cel.glmm.fit = glmer(multnom_out ~ nt1v + (1|t_number), data = recgli[which(recgli$in_CEL ==1),], family="binomial")
nt1v.nel.glmm.fit = glmer(multnom_out ~ nt1v + (1|t_number), data = recgli[which(recgli$in_NEL ==1),], family="binomial")
print("In CEL:")
round(summary(nt1v.cel.glmm.fit)$coefficients, 4)
print("In NEL:")
round(summary(nt1v.nel.glmm.fit)$coefficients, 4)
```

### f. Use GEE to associate anatomical parameters with TxE/rHGG, separating into CEL & NEL 

```r
# NFSE
nfse.gee.cel.fit = geeglm(binom_out ~ nfse, data = recgli[which(recgli$in_CEL ==1),], id = factor(t_number), family = binomial, corstr = "exchangeable")
nfse.gee.nel.fit = geeglm(binom_out ~ nfse, data = recgli[which(recgli$in_NEL ==1),], id = factor(t_number), family = binomial, corstr = "exchangeable")
print("In CEL:")
round(summary(nfse.gee.cel.fit)$coefficients, 4)
print("In NEL:")
round(summary(nfse.gee.nel.fit)$coefficients, 4)
```

```r
# NFL
nfl.gee.cel.fit = geeglm(binom_out ~ nfl, data = recgli[which(recgli$in_CEL ==1),], id = factor(t_number), family = binomial, corstr = "exchangeable")
nfl.gee.nel.fit = geeglm(binom_out ~ nfl, data = recgli[which(recgli$in_NEL ==1),], id = factor(t_number), family = binomial, corstr = "exchangeable")
print("nFSE in CEL:")
round(summary(nfl.gee.cel.fit)$coefficients, 4)
print("nFSE in NEL:")
round(summary(nfl.gee.nel.fit)$coefficients, 4)
```

```r
# NT1c
nt1c.gee.cel.fit = geeglm(binom_out ~ nt1c, data = recgli[which(recgli$in_CEL ==1),], id = factor(t_number), family = binomial, corstr = "exchangeable")
nt1c.gee.nel.fit = geeglm(binom_out ~ nt1c, data = recgli[which(recgli$in_NEL ==1),], id = factor(t_number), family = binomial, corstr = "exchangeable")
print("nT1c in CEL:")
round(summary(nt1c.gee.cel.fit)$coefficients, 4)
print("nT1c in NEL:")
round(summary(nt1c.gee.nel.fit)$coefficients, 4)
```

```r
# NT1v
nt1v.gee.cel.fit = geeglm(binom_out ~ nt1v, data = recgli[which(recgli$in_CEL ==1 & !is.na(recgli$nt1v)),], id = factor(t_number), family = binomial, corstr = "exchangeable")
nt1v.gee.nel.fit = geeglm(binom_out ~ nt1v, data = recgli[which(recgli$in_NEL ==1 & !is.na(recgli$nt1v)),], id = factor(t_number), family = binomial, corstr = "exchangeable")
print("nT1v in CEL:")
round(summary(nt1v.gee.cel.fit)$coefficients, 4)
print("nT1v in NEL:")
round(summary(nt1v.gee.nel.fit)$coefficients, 4)
```

## 2. Combining anatomical parameters to look at different associations 

### a. Using GLMM: 

```r
anat.glmm.fit = glmer(multnom_out ~ nfse + nfl + nt1c + nt1v + (1|t_number), data = recgli, family="binomial")
round(summary(anat.glmm.fit)$coefficients, 4) 
```

### b. Using GEE:  

```r
anat.gee.fit = geeglm(binom_out ~ nfse + nfl + nt1c+ nt1v, data = recgli[!is.na(recgli$nt1v),], id = factor(t_number), family = binomial, corstr = "exchangeable")
round(summary(anat.gee.fit)$coefficients, 4) 
```

## 3. Using in_CEL as a covariate in anat params

### a. Using GLMM: 

```r
anat.glmm.cel.fit = glmer(multnom_out ~ nfse*in_CEL + (1|t_number), data = recgli, family="binomial")
round(summary(anat.glmm.cel.fit)$coefficients, 4) 
```

```r
anat.glmm.cel.fit = glmer(multnom_out ~ nfl*in_CEL + (1|t_number), data = recgli, family="binomial")
round(summary(anat.glmm.cel.fit)$coefficients, 4) 
```

```r
anat.glmm.cel.fit = glmer(multnom_out ~ nt1c*in_CEL + (1|t_number), data = recgli, family="binomial")
round(summary(anat.glmm.cel.fit)$coefficients, 4) 
```

```r
anat.glmm.cel.fit = glmer(multnom_out ~ nt1v*in_CEL + (1|t_number), data = recgli, family="binomial")
round(summary(anat.glmm.cel.fit)$coefficients, 4) 
```

```r
anat.glmm.cel.fit = glmer(multnom_out ~ nfse+nfl+nt1c+nt1v*in_CEL + (1|t_number), data = recgli, family="binomial")
round(summary(anat.glmm.cel.fit)$coefficients, 4) 
## Note: this is the only one where we can use in_CEL as an interaction term & we get convergence; 
## everything else (interaction with other variables) does not converge.
```

### b. Using GEE: 

```r
anat.gee.cel.fit = geeglm(binom_out ~ nfse*in_CEL , data = recgli[!is.na(recgli$nt1v),], id = factor(t_number), family = binomial, corstr = "exchangeable")
round(summary(anat.gee.cel.fit)$coefficients, 4) 
```

```r
anat.gee.cel.fit = geeglm(binom_out ~ nfl*in_CEL , data = recgli[!is.na(recgli$nt1v),], id = factor(t_number), family = binomial, corstr = "exchangeable")
round(summary(anat.gee.cel.fit)$coefficients, 4) 
```

```r
anat.gee.cel.fit = geeglm(binom_out ~ nt1c*in_CEL , data = recgli[!is.na(recgli$nt1v),], id = factor(t_number), family = binomial, corstr = "exchangeable")
round(summary(anat.gee.cel.fit)$coefficients, 4) 
```

```r
anat.gee.cel.fit = geeglm(binom_out ~ nt1v*in_CEL , data = recgli[!is.na(recgli$nt1v),], id = factor(t_number), family = binomial, corstr = "exchangeable")
round(summary(anat.gee.cel.fit)$coefficients, 4) 
```

```r
anat.gee.cel.fit = geeglm(binom_out ~ nfse+nfl+nt1c+nt1v+in_CEL , data = recgli[!is.na(recgli$nt1v),], id = factor(t_number), family = binomial, corstr = "exchangeable")
round(summary(anat.gee.cel.fit)$coefficients, 4) 
```
