```r
table(recgli$multnom_out)
table(is.na(recgli$multnom_out))
```

## 1. Diffusion vs. rHGG/TxE

### a. No separation into ROIs

```r
par(mfrow = c(2, 2))
a = boxplot(nadc.1 ~ multnom_out , data = recgli, ylab = "normalized ADC", main = "median ADC")
text( c(1:nlevels(factor(recgli$multnom_out))) , a$stats[nrow(a$stats) , ]+0.5 , paste("n = ",table(recgli$multnom_out),sep="")  )

b = boxplot(nfa.1 ~ multnom_out , data = recgli, ylab = "normalized FA", main = "median FA")
text( c(1:nlevels(factor(recgli$tumor_cell_evaluation))) , b$stats[nrow(b$stats) , ]+0.5 , paste("n = ",table(recgli$multnom_out),sep="")  )

c = boxplot(nev1.1 ~ multnom_out , data = recgli, ylab = "normalized EV1", main = "median EV1")
text( c(1:nlevels(factor(recgli$multnom_out))) , c$stats[nrow(c$stats) , ]+0.5 , paste("n = ",table(recgli$multnom_out),sep="")  )

d = boxplot(nevrad.1 ~ multnom_out , data = recgli, ylab = "normalized EVRAD", main = "median EVRAD")
text( c(1:nlevels(factor(recgli$multnom_out))) , d$stats[nrow(d$stats) , ]+0.5 , paste("n = ",table(recgli$multnom_out),sep="")  )
```

### b. CEL next to NEL scores: 

```r
par(mfrow = c(2, 2))
a = boxplot(nadc.1 ~ multnom_out , data = recgli[recgli$in_CEL ==1,], ylab = "normalized nadc.1", main = "CEL median nadc.1", ylim = c(0,4))
text( c(1:nlevels(factor(recgli$multnom_out))) , a$stats[nrow(a$stats) , ]+.3 , paste("n = ",table(recgli[which(recgli$in_CEL ==1),]$multnom_out),sep=""), cex = 0.8)
b = boxplot(nadc.1 ~ multnom_out , data = recgli[recgli$in_NEL ==1,], ylab = "normalized nadc.1", main = "NEL median nadc.1", ylim = c(0,4))
text( c(1:nlevels(factor(recgli$multnom_out))) , b$stats[nrow(b$stats) , ]+.3 , paste("n = ",table(recgli[which(recgli$in_NEL ==1),]$multnom_out),sep=""), cex = 0.8)

c = boxplot(nfa.1 ~ multnom_out , data = recgli[recgli$in_CEL ==1,], ylab = "normalized nfa.1", main = "CEL median nfa.1", ylim = c(0,3.5))
text( c(1:nlevels(factor(recgli$multnom_out))) , c$stats[nrow(c$stats) , ]+.3 , paste("n = ",table(recgli[which(recgli$in_CEL ==1),]$multnom_out),sep=""), cex = 0.8)
d = boxplot(nfa.1 ~ multnom_out , data = recgli[recgli$in_NEL ==1,], ylab = "normalized nfa.1", main = "NEL median nfa.1", ylim = c(0,3.5))
text( c(1:nlevels(factor(recgli$multnom_out))) , d$stats[nrow(d$stats) , ]+.3 , paste("n = ",table(recgli[which(recgli$in_NEL ==1),]$multnom_out),sep=""), cex = 0.8)

e = boxplot(nev1.1 ~ multnom_out , data = recgli[recgli$in_CEL ==1,], ylab = "normalized nev1.1", main = "CEL median nev1.1", ylim = c(0,4))
text( c(1:nlevels(factor(recgli$multnom_out))) , e$stats[nrow(e$stats) , ]+.3 , paste("n = ",table(recgli[recgli$in_CEL ==1,]$multnom_out),sep=""), cex = 0.8)
f = boxplot(nev1.1 ~ multnom_out , data = recgli[recgli$in_NEL ==1,], ylab = "normalized nev1.1", main = "NEL median nev1.1", ylim = c(0,4))
text( c(1:nlevels(factor(recgli$multnom_out))) , f$stats[nrow(f$stats) , ]+.3 , paste("n = ",table(recgli[recgli$in_NEL ==1,]$multnom_out),sep=""), cex = 0.8)

g = boxplot(nevrad.1 ~ multnom_out , data = recgli[recgli$in_CEL ==1,], ylab = "normalized nevrad.1", main = "CEL median nevrad.1", ylim = c(0,4))
text( c(1:nlevels(factor(recgli$multnom_out))) , g$stats[nrow(g$stats) , ]+.3 , paste("n = ",table(recgli[recgli$in_CEL ==1,]$multnom_out),sep=""), cex = 0.8)
h = boxplot(nevrad.1 ~ multnom_out , data = recgli[recgli$in_NEL ==1,], ylab = "normalized nevrad.1", main = "NEL median nevrad.1", ylim = c(0,4))
text( c(1:nlevels(factor(recgli$multnom_out))) , h$stats[nrow(h$stats) , ]+.3 , paste("n = ",table(recgli[recgli$in_NEL ==1,]$multnom_out),sep=""), cex = 0.8)
```

### d. Use GEE to associate diffusion parameters with rHGG/TxE, no separation


```r
## here we have to code a binomial outcome rather than using factor TxE and rHGG for some reason: 
recgli = data.frame(recgli, binom_out = ifelse(recgli$multnom_out == "rHGG", 1, 0))
recgli_nadc = recgli[!is.na(recgli$nadc.1),]
recgli_nfa = recgli[!is.na(recgli$nfa.1),]
recgli_nev1 = recgli[!is.na(recgli$nev1.1),]
recgli_nevrad = recgli[!is.na(recgli$nevrad.1),]
```

```r
nadc.gee.fit = geeglm(binom_out ~ nadc.1, data = recgli_nadc, id = factor(t_number), family = binomial, corstr = "exchangeable")
round(summary(nadc.gee.fit)$coefficients, 4)
```

```r
nfa.gee.fit = geeglm(binom_out ~ nfa.1, data = recgli_nfa, id = factor(t_number), family = binomial, corstr = "exchangeable")
round(summary(nfa.gee.fit)$coefficients, 4)
```

```r
nev1.gee.fit = geeglm(binom_out ~ nev1.1, data = recgli_nev1, id = factor(t_number), family = binomial, corstr = "exchangeable")
round(summary(nev1.gee.fit)$coefficients, 4)
```

```r
nevrad.gee.fit = geeglm(binom_out ~ nevrad.1, data = recgli_nevrad, id = factor(t_number), family = binomial, corstr = "exchangeable")
round(summary(nevrad.gee.fit)$coefficients, 4)
```

### e. Use GLMM to associate diffusion parameters with TxE/rHGG, separating into CEL & NEL 

```r
# ADC 
nadc.cel.glmm.fit = glmer(multnom_out ~ nadc.1 + (1|t_number), data = recgli[which(recgli$in_CEL ==1),], family="binomial")
nadc.nel.glmm.fit = glmer(multnom_out ~ nadc.1 + (1|t_number), data = recgli[which(recgli$in_NEL ==1),], family="binomial")
print("In CEL:")
round(summary(nadc.cel.glmm.fit)$coefficients, 4)
print("In NEL:")
round(summary(nadc.nel.glmm.fit)$coefficients, 4)
```

```r
# FA 
nfa.cel.glmm.fit = glmer(multnom_out ~ nfa.1 + (1|t_number), data = recgli[which(recgli$in_CEL ==1),], family="binomial")
nfa.nel.glmm.fit = glmer(multnom_out ~ nfa.1 + (1|t_number), data = recgli[which(recgli$in_NEL ==1),], family="binomial")
print("In CEL:")
round(summary(nfa.cel.glmm.fit)$coefficients, 4)
print("In NEL:")
round(summary(nfa.nel.glmm.fit)$coefficients, 4)
```

```r
# EV1
nev1.cel.glmm.fit = glmer(multnom_out ~ nev1.1 + (1|t_number), data = recgli[which(recgli$in_CEL ==1),], family="binomial")
nev1.nel.glmm.fit = glmer(multnom_out ~ nev1.1 + (1|t_number), data = recgli[which(recgli$in_NEL ==1),], family="binomial")
print("In CEL:")
round(summary(nev1.cel.glmm.fit)$coefficients, 4)
print("In NEL:")
round(summary(nev1.nel.glmm.fit)$coefficients, 4)
```

```r
# EVRAD
nevrad.cel.glmm.fit = glmer(multnom_out ~ nevrad.1 + (1|t_number), data = recgli[which(recgli$in_CEL ==1),], family="binomial")
nevrad.nel.glmm.fit = glmer(multnom_out ~ nevrad.1 + (1|t_number), data = recgli[which(recgli$in_NEL ==1),], family="binomial")
print("In CEL:")
round(summary(nevrad.cel.glmm.fit)$coefficients, 4)
print("In NEL:")
round(summary(nevrad.nel.glmm.fit)$coefficients, 4)
```

### f. Use GEE to associate anatomical parameters with TxE/rHGG, separating into CEL & NEL

```r
# NFA
nfa.gee.cel.fit = geeglm(binom_out ~ nfa.1, data = recgli_nfa[which(recgli_nfa$in_CEL ==1),], id = factor(t_number), family = binomial, corstr = "exchangeable")
nfa.gee.nel.fit = geeglm(binom_out ~ nfa.1, data = recgli_nfa[which(recgli_nfa$in_NEL ==1),], id = factor(t_number), family = binomial, corstr = "exchangeable")
print("In CEL:")
round(summary(nfa.gee.cel.fit)$coefficients, 4)
print("In NEL:")
round(summary(nfa.gee.nel.fit)$coefficients, 4)
```

```r
# NEV1
nev1.gee.cel.fit = geeglm(binom_out ~ nev1.1, data = recgli_nev1[which(recgli_nev1$in_CEL ==1),], id = factor(t_number), family = binomial, corstr = "exchangeable")
nev1.gee.nel.fit = geeglm(binom_out ~ nev1.1, data = recgli_nev1[which(recgli_nev1$in_NEL ==1),], id = factor(t_number), family = binomial, corstr = "exchangeable")
print("In CEL:")
round(summary(nev1.gee.cel.fit)$coefficients, 4)
print("In NEL:")
round(summary(nev1.gee.nel.fit)$coefficients, 4)
```

```r
# NEVRAD
nevrad.gee.cel.fit = geeglm(binom_out ~ nevrad.1, data = recgli_nevrad[which(recgli_nevrad$in_CEL ==1),], id = factor(t_number), family = binomial, corstr = "exchangeable")
nevrad.gee.nel.fit = geeglm(binom_out ~ nevrad.1, data = recgli_nevrad[which(recgli_nevrad$in_NEL ==1),], id = factor(t_number), family = binomial, corstr = "exchangeable")
print("In CEL:")
round(summary(nevrad.gee.cel.fit)$coefficients, 4)
print("In NEL:")
round(summary(nevrad.gee.nel.fit)$coefficients, 4)
```

## 2. Combining diffusion parameters to look at different associations

### b. Using GEE:  

```r
diffu.gee.fit = geeglm(binom_out ~ nadc.1+nfa.1+nev1.1+nevrad.1, data = recgli_nevrad, id = factor(t_number), family = binomial, corstr = "exchangeable")
round(summary(diffu.gee.fit)$coefficients, 4) 
```
